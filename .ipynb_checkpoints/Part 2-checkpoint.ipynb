```python
import numpy as np
import pandas as pd
from random import  uniform
import statistics
import matplotlib.pyplot as plt
```

# Rosenblatt neuron class

```python
class Neuron:
    
    def __init__(self):
        self.data= np.array([])
        self.load_data()
        self.weights = np.array([
            uniform(-.100, .100),
            uniform(-1, 1),
            uniform(-.5, .5)
        ])
        
        self.teta = 0
        self.ta   = 0.1
        self.vies = -1
        self.total_of_adjustments = []
        self.epoch = 0
        self.x1 = 0
        self.x2 = 0
    
    def changes_initial_weights(self,x1,x2,x3):
        self.weights[0] = self.weights[0]*x1
        self.weights[1] = self.weights[1]*x2
        self.weights[2] = self.weights[2]*x3

    def load_data(self):
        dataset = np.fromfile('data1.txt')
        rows  = len(dataset)//3
        columns = 3
        self.data=dataset.reshape(rows, columns)
    
    def show_graphic(self):
        fig, ax = plt.subplots()
        for i in self.data:
            x = i[0]
            y = i[1]
            yd = int(i[2])
            if(yd >= 1):
                plt.scatter(x,y, color='r', s=8)
            else:
                plt.scatter(x,y, color='b', s=8)

        point1 = [0,self.x1]
        point2 = [self.x2,0]
        x_values = [point1[0], point2[0]]
        y_values = [point1[1], point2[1]]

        ax.plot(x_values, y_values)
        plt.show()
        
    def function_activation(self,num):
        if (num>=self.teta):
            return 1
        else:
            return 0
        
    def perceptron(self):
        adjustment_number = 0
        
        for i in self.data:
            input_value = np.concatenate((np.array([self.vies]), i[:2]))
            yd = int(i[2:])
            yr = round(input_value.dot(self.weights),4)
            yr = self.function_activation(yr)
            
            adjustment_number += 1
            
            self.weights = self.weight_adjustment(yd, yr, (input_value))
            
            if(yr != yd):
                self.total_of_adjustments.append(adjustment_number)
                return False
            
        self.total_of_adjustments.append(adjustment_number)
        return True

    def weight_adjustment(self, yd, yr, input_value):
        delta = self.ta * np.array(yd - yr).dot((input_value))
        novos_weights = self.weights + delta
        
        return novos_weights
    
    def find_x1_x2(self):
        w = self.weights
        
        w0_w2 = w[0]/w[2]
        w1_w2 = w[1]/w[2]

        self.x2 = w0_w2 - w1_w2 * 0
        self.x1 = w0_w2/w1_w2
    
    def training(self):
        self.epoch = 1
        #100 interações
        for i in range(100):
            if(self.perceptron() == False):
                self.epoch += 1
            else:
                break
            
        self.find_x1_x2()
        
        print("Total epoch: ",self.epoch)
        print("Total adjustments made to weights: ",sum(self.total_of_adjustments))
        print("Average adjustments made to weights: ",sum(self.total_of_adjustments)/len(self.total_of_adjustments))
        print("Standard deviation of adjustments made to weights: ",statistics.stdev(self.total_of_adjustments))
        
        return self.epoch, sum(self.total_of_adjustments), sum(self.total_of_adjustments)/len(self.total_of_adjustments), statistics.stdev(self.total_of_adjustments)
        
n1 = Neuron()
n2 = Neuron()
n3 = Neuron()
n4 = Neuron()
n5 = Neuron()
n6 = Neuron()
n7 = Neuron()
n8 = Neuron()
n9 = Neuron()
```

# Table with comparison of the mean and standard deviation of the 9 configurations

```python
from prettytable import PrettyTable
dat_dtype = {
    'names' : ('configuration', 'epochs', 'total adjustments', 'average adjustments', 'standard deviation'),
    'formats' : ('i', 'i', 'i', 'd', 'd')}

dat = np.zeros(9, dat_dtype)

f = [1, 2, 3, 4, 5, 6, 7, 8, 9]
i = 0

while i < 9:
    dat['configuration'][i] = f[i]
    dat['epochs'][i] = epochs[i]
    dat['total adjustments'][i] = total_adjustments[i]
    dat['average adjustments'][i] = average_adjustments[i]
    dat['standard deviation'][i] = standard_deviation[i]
    
    i+=1

x = PrettyTable(dat.dtype.names)

for row in dat:
    x.add_row(row)
```

```python
print(x)
```

From the Table we can see that configuration 7 had the least number of times, the lowest total weight adjustments, the lowest average weight adjustments and the largest standard deviation.
